```python
import pandas as pd
from sqlalchemy import create_engine
from urllib.parse import quote_plus

password = quote_plus("Judah@25.")

engine = create_engine(
    f"mysql+mysqlconnector://root:{password}@localhost:3306/retail_analytics_db"
)

query = """
SELECT 
    o.order_id,
    o.order_date,
    c.customer_id,
    s.store_name,
    s.region,
    SUM(oi.quantity * oi.unit_price) AS total_sales
FROM orders o
JOIN order_items oi ON o.order_id = oi.order_id
JOIN customers c ON o.customer_id = c.customer_id
JOIN stores s ON o.store_id = s.store_id
GROUP BY o.order_id, o.order_date, c.customer_id, s.store_name, s.region;
"""

orders_df = pd.read_sql(query, engine)
orders_df.head()
```
