```python
import pandas as pd

orders_df = pd.read_csv("../Python_Analysis/data/orders.csv")

orders_df.head()

customer_metrics = (
    orders_df
    .groupby('customer_id')
    .agg(
        total_orders=('order_id', 'nunique'),
        total_revenue=('total_sales', 'sum'),
        avg_order_value=('total_sales', 'mean'),
        last_purchase=('order_date', 'max')
    )
    .reset_index()
)

customer_metrics
```
