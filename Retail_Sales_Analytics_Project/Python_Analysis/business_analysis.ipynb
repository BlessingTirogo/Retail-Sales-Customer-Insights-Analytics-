```python
import pandas as pd

orders_df = pd.read_csv("../Python_Analysis/data/orders.csv")

orders_df.head()
```

```python
#Monthly Revenue Trend
orders_df['order_date'] = pd.to_datetime(orders_df['order_date'])

monthly_revenue = (
    orders_df
    .groupby(orders_df['order_date'].dt.to_period('M'))
    ['total_sales']
    .sum()
    .reset_index()
)

monthly_revenue
```

```python
#Top Customers by Revenue

top_customers = (
    orders_df
    .groupby('store_name')['total_sales']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

top_customers
```

```python
#Regional performance
region_revenue = (
    orders_df
    .groupby('region')['total_sales']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

region_revenue
```

```python
#Prepare Monthly Revenue Properly
orders_df['order_date'] = pd.to_datetime(orders_df['order_date'])

monthly_revenue = (
    orders_df
    .groupby(orders_df['order_date'].dt.to_period('M'))
    ['total_sales']
    .sum()
    .reset_index()
)

monthly_revenue['order_month'] = monthly_revenue['order_date'].astype(str)
monthly_revenue
```

```python
#Month-over-Month (MoM) Growth
monthly_revenue = monthly_revenue.sort_values('order_month')

monthly_revenue['previous_month_sales'] = (
    monthly_revenue['total_sales'].shift(1)
)

monthly_revenue['mom_growth_pct'] = (
    (monthly_revenue['total_sales'] - monthly_revenue['previous_month_sales'])
    / monthly_revenue['previous_month_sales']
) * 100

monthly_revenue
```

```python
#Regional Performance Analysis
regional_sales = (
    orders_df
    .groupby('region')['total_sales']
    .sum()
    .reset_index()
    .sort_values('total_sales', ascending=False)
)

regional_sales
```

```python
#Store Performance (Top & Bottom Stores)
store_performance = (
    orders_df
    .groupby('store_name')['total_sales']
    .sum()
    .reset_index()
    .sort_values('total_sales', ascending=False)
)

store_performance.head(5)   # Top stores
store_performance.tail(5)   # Bottom stores
```

```python
import os
os.makedirs("../data/processed", exist_ok=True)

monthly_revenue.to_csv(
    "../data/processed/monthly_revenue.csv", index=False
)

regional_sales.to_csv(
    "../data/processed/regional_sales.csv", index=False
)

store_performance.to_csv(
    "../data/processed/store_performance.csv", index=False
)
```
